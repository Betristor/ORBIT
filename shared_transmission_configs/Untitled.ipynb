```python
from ORBIT.phases.design import ElectricalDesign
from ORBIT import ParametricManager, ProjectManager
import yaml
import pandas as pd
import numpy as np
from copy import deepcopy
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
df = pd.read_excel("C:/Users/sbredenk/ORBIT/osw_project_details.xlsx",sheet_name=0)
proj_name = df['name'].to_list()
filename = deepcopy(proj_name)
for i in range(0,len(filename)):
    filename[i] = "%s base.yaml" % proj_name[i]
```

```python
total_capex = {}
for i in range(0,len(proj_name)):
    with open(filename[i], "r") as fh:
        config = yaml.load(fh, Loader=yaml.SafeLoader)
    project = ProjectManager(config)
    project.run()
    total_capex[proj_name[i]] = project.capex_breakdown
# print(total_capex)
```
